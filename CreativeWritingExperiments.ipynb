#**Standard IO**

```python
import argparse
from tot.methods.bfs import solve, naive_solve
from tot.tasks.text import TextTask

args = argparse.Namespace(
    backend='gpt-4',
    temperature=0.7,
    task='text',
    naive_run=True,  # Add this for standard IO
    prompt_sample='standard',  # standard IO prompt
    method_generate='sample',
    method_evaluate='vote',
    method_select='greedy',
    n_generate_sample=1,
    n_evaluate_sample=3,
    n_select_sample=5
)

task = TextTask()

# Try first 20 problems
for idx in range(41, 61):
    print(f"\n=== Problem {idx} ===")
    ys, infos = naive_solve(args, task, idx)  # Use naive_solve instead of solve
    print(f"\nAnswer:")
    print(ys[0])
    print("-" * 50)
```
